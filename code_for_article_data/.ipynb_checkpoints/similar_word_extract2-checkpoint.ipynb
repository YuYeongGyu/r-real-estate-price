```python
case_index_df = pd.read_pickle("../seoul_apt_trade_data/case_index_df.pkl")
case_index_df
```

```python
gu_name = ['강남구','강동구','강서구','강북구','관악구','광진구','구로구','금천구',
          '노원구','동대문구','도봉구','동작구','마포구','서대문구','성동구','성북구',
          '서초구','송파구','영등포구','용산구','양천구','은평구','종로구','중구','중랑구']
```

```python
trade_data = {}
for gu in gu_name:
    df = pd.read_pickle(f"../seoul_apt_trade_data/{gu}_trade.pkl")
    df = df.groupby(['거래주']).mean()['면적당가격']
    for i in range(808):
        if i not in df.index:
            df.iloc[i] = df.iloc[i-1]
    trade_data[gu] = df 
```

```python
total_dict = {}
for gu in case_index_df.index:
    print(f'>>> start {gu}')
    gu_case_dict = {}
    for i in range(1,9):
        case = f'{i}case'
        case_dict = {}
        case_list = case_index_df.loc[gu,f'case{i}']
        for idx in case_list:
            word_list = ft_idf.iloc[idx,:].sort_values(ascending = False).index
            case_dict[idx] = word_list
        gu_case_dict[case] = case_dict
    total_dict[gu] = gu_case_dict
    print(f'>>> finish {gu}')
```

```python
df = pd.DataFrame(total_dict['강남구']['1case'])
```

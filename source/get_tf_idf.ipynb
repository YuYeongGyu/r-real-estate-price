# Info
: 전체 기사 title word들에 대한 tf-idf 값을 구하기 위한 코드
한 주를 doc으로 한 tf-idf 값과 한 달을 doc으로 한 tf-idf 값을 pickle 파일로 저장

```python
import pandas as pd
import glob

file_name = glob.glob('../article/headline_noun_keyword*')

keyword_df = []
for file in file_name:
    df = pd.read_pickle(file)
    keyword_df.append(df)

keyword_all = pd.concat(keyword_df, ignore_index = True)
keyword = keyword_all[keyword_all['week'] <= 807]
keyword
```

주간 TF-IDF값 구하기

```python
import pos_to_dict

weekly_df_all = pos_to_dict.weekly_pos_to_dict(keyword)
weekly_df_all
```

```python
week = len(weekly_df_all.index)
```

```python
import math

idf_list = []
for i in weekly_df_all.index:
    idf = 0
    for j in weekly_df_all.loc[i,:]:
        if j != 0:
            idf += 1
    idf_list.append(math.log(week/idf))
```

```python
tf = pd.DataFrame()

max_list = []
for i in weekly_df_all.columns[:-1]:
    _max = max(list(weekly_df_all.loc[:,i]))
    tf[i] = weekly_df_all.loc[:,i]*0.5/_max + 0.5
```

```python
ft_idf = pd.DataFrame()

idx = 0
for i in weekly_df_all.index:
    ft_idf_list = weekly_df_all.loc[i,:]*idf_list[idx]
    ft_idf[i] = ft_idf_list
    idx += 1
    if idx%1000 == 0:
        print(f">>> get {idx} words TF-IDF finished")
```

```python
ft_idf.to_pickle("../article/tf_idf.pkl")
```

월간 TF-IDF값 구하기

```python
file_name = glob.glob('../article/headline_noun_keyword*')

keyword_df = []
for file in file_name:
    df = pd.read_pickle(file)
    keyword_df.append(df)

keyword_all = pd.concat(keyword_df, ignore_index = True)
keyword = keyword_all[keyword_all['week'] <= 807]
keyword
```

```python
def cal_month(date):
    year = int(date[0])
    month = int(date[1])
    return (year-2006)*12+month-1
```

```python
keyword['month'] = [cal_month(date[:10].split('.')) for date in keyword['date']]

monthly_pos = keyword.groupby('month')['pos'].sum()

monthly_word_dict = {}

for m in range(len(monthly_pos)):
    word_dict = {}
    for word in monthly_pos[m]:
        if word not in word_dict:
            word_dict[word] = 1
        else:
            word_dict[word] += 1
    monthly_word_dict[m] = word_dict
monthly_word_df = pd.DataFrame(monthly_word_dict)
monthly_word_df = monthly_word_df.fillna(0)
monthly_word_df
```

```python
month = len(monthly_word_df.columns)
month
```

```python
import math

idf_list = []
for i in monthly_word_df.index:
    idf = 0
    for j in monthly_word_df.loc[i,:]:
        if j != 0:
            idf += 1
    idf_list.append(math.log(month/idf))
```

```python
tf = pd.DataFrame()

max_list = []
for i in monthly_word_df.columns:
    _max = max(list(monthly_word_df.loc[:,i]))
    tf[i] = monthly_word_df.loc[:,i]*0.5/_max + 0.5
```

```python
ft_idf_month = pd.DataFrame()

idx = 0
for i in monthly_word_df.index:
    ft_idf_list = monthly_word_df.loc[i,:]*idf_list[idx]
    ft_idf_month[i] = ft_idf_list
    idx += 1
    if idx%1000 == 0:
        print(f">>> get {idx} words TF-IDF finished")
```

```python
ft_idf_month.to_pickle("../article/tf_idf_month.pkl")
```

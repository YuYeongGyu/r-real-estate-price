```python
import pandas as pd

read_path = '../article/headline_keyword/'
table = pd.read_csv(f'{read_path}keyword_about_location.csv')

del table['Unnamed: 0']
```

```python
gu_name = ['강남구','강동구','강서구','강북구','관악구','광진구','구로구','금천구',
          '노원구','동대문구','도봉구','동작구','마포구','서대문구','성동구','성북구',
          '서초구','송파구','영등포구','용산구','양천구','은평구','종로구','중구','중랑구']

many = len(gu_name)
```

```python
df_list = []
for name in gu_name:
    df = pd.DataFrame()
    for week in range(783):
        sub_df = pd.read_csv(f'../seoul_apt_trade_data/weekly_trade/{name}_{week}week_trade.csv',encoding = 'cp949')
        del sub_df['Unnamed: 0']
        del sub_df['Unnamed: 0.1']
        df = pd.concat([df,sub_df],ignore_index=True)
    df_list.append(df)
```

```python
df_list[0]
```

```python
import matplotlib.pyplot as plt
x=df_list[0].groupby('거래주').mean()['면적당가격']
y1=table['서울']
plt.scatter(x,y1)
```

```python
y2=table['강남']
plt.scatter(x,y2)
```

```python
from scipy import stats
print(stats.pearsonr(x,y1))
print(stats.spearmanr(x,y1))
```

```python
print(stats.pearsonr(x,y2))
print(stats.spearmanr(x,y2))
```
